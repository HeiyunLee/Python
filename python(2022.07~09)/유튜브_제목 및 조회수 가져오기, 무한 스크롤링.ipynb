```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as req
import time
from tqdm import tqdm_notebook
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get("https://www.youtube.com/c/KakaoFriendsofficial/videos")
```

```python
body = driver.find_element_by_css_selector('body')
body.send_keys(Keys.END)
```

```python
soup = bs(driver.page_source, "lxml")
```

```python
body = driver.find_element_by_css_selector('body')

while True:
    last = bs(driver.page_source, "lxml")
    
    for i in range(10):
        body.send_keys(Keys.END)
    
    time.sleep(1)
    
    new = bs(driver.page_source, "lxml")
    
    if last == new:
        break
    else:
        pass
```

```python
video_title = driver.find_elements_by_css_selector('#video-title')
video_title
```

```python
video_title_list = []
for i in video_title:
    video_title_list.append(i.text)
video_title_list
```

```python
renderer = driver.find_elements_by_css_selector('#metadata-line > span:nth-child(1)')
renderer
```

```python
renderer_list = []
for i in renderer:
    renderer_list.append(i.text)
renderer_list
```

```python
len(video_title_list), len(renderer_list)
```

```python
data = {'제목': video_title_list, '조회수': renderer_list}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.set_index('조회수')
df
```
