```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', marker = '*', lw = 5, color = 'y', 
         ms = '15', markerfacecolor = 'w', markeredgecolor = 'c', markeredgewidth = 2)
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', marker = '<', lw = 2, color = 'm', 
         ms = '10', markerfacecolor = 'k', markeredgecolor = 'k', markeredgewidth = 2)
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', marker = '<', lw = 2, color = 'm', 
         ms = '10', markerfacecolor = 'k', markeredgecolor = 'k', markeredgewidth = 2)
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, ls = '-.', marker = '<', lw = 2, color = 'm', 
         ms = '10', markerfacecolor = 'k', markeredgecolor = 'k', markeredgewidth = 2)
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.grid()
plt.show()
```

```python
plt.plot(x,y,ls = "--", lw = 5, marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
# 오류: 오타를 확인하자..!!!!
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.show()
```

```python
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.legend()
# legend: 범례 설정
plt.show()
```

```python
plt.plot(x, label = 'x축')
plt.plot(y, label = 'y축')
plt.plot(z, label = 'z축')
plt.legend()
plt.show()
```
