```python
s4 = pd.read_csv('2014년 졸음운전 교통사고.csv', encoding = 'euc-kr')
s5 = pd.read_csv('2015년 졸음운전 교통사고.csv', encoding = 'euc-kr')
s6 = pd.read_csv('2016년 졸음운전 교통사고.csv', encoding = 'euc-kr')
# 오류: 괄호가 빠져서 신텍스 에러남..

print(s4, s5, s6)
```

```python
data = pd.concat([s4, s5, s6])
data.set_index('구분', inplace = True)
data
# 오류: 위에서 '구분'으로 먼저 인덱싱을 해뒀더니 출력이 되지 않았다.
```

```python
ans2 = data.copy()
ans2.index = pd.to_datetime(ans2.index, format = '%Y년%m월')
# 문자열의 데이타를 숫자로 바꾼다.
ans2.set_index([ans2.index.year,ans2.index.month], inplace = True)
ans2.index.names = ['년도', '월']
ans2
```

```python
y = ans2.groupby('년도').mean()[['사망(명)']]
# 변수.groupby(기준으로 보고 싶은 열).mean()[[원하는 기준값]]
y
```

```python
m = ans2.groupby('월').mean()[['사망(명)']]
m
```

```python
ans2.loc[2016]
# ans2: 카피본..
# loc : 인덱싱 이름 가져오는 함수
```

```python
# 2016년도 사고 대비 사망률
dead_rate = ans2.loc[2016]['사망(명)'].sum() / ans2.loc[2016]['사고(건)'].sum() * 100
dead_rate
```

```python
ans2.loc[2016]['사망(명)'].sum() / ans2.loc[2016]['사고(건)'].sum() * 100
# loc를 사용해 2016을 뽑고 2016에서 '사망(명)'이 라는 칼럼을 가져온다.
# 변수 선언 안해도 출략 가능..
```

```python
d_2014 = ans2.loc[2014]
d_2014
```

```python
ans2.loc[2014][['사망(명)','부상(명)']].plot(kind = 'bar', rot = 0)
plt.figure(figsize = (20,10))
# plt.figure(figsize = (가로길이, 세로길이))
plt.show()
```

```python
plt.style.use('ggplot')
# plt.style.use('ggplot'): 그리드가 들어간 그래프 출력
d_2014[['사망(명)', '부상(명)']].plot(kind = 'bar', rot = 0)
# rot: x축 라벨의 회전(ratation) 각도를 설정하는 함수.
# rot을 0으로 설정하지 않으면 글자가 누워있음..

plt.title('2014년 월별 사망자와 부상자 수')
plt.show()
```
