```python
# 전체 정리
```

```python
title_list = []
rating_list = []

while True:
    try:
        for i in range(1,6):
            page = driver.find_element_by_xpath('//*[@id="info.search.page.no{}"]'.format(i))
            page.click()

            soup = bs(driver.page_source, 'lxml')
            time.sleep(1)

            title = soup.select('a.link_name')
            rating = soup.select('span.score > em')

            for j in range(len(title)):
                title_list.append(title[j].text)
                rating_list.append(rating[j].text)
                print(title[j].text, rating[j].text)
        
            time.sleep(1)
        
    # 다음 버튼 (> 표시)
        button = driver.find_element_by_css_selector('#info\.search\.page\.next')
        button.click() 
    
    except:
        break
```

```python
data = {'가게이름' : title_list, '평점' : rating_list}
data
```

```python
restarant = pd.DataFrame(data)
restarant
```

```python
restarant = restarant.set_index('평점')
restarant
```

```python
restarant_acs = restarant.sort_index(ascending = False)
restarant_acs
```
