```python
def extract_pos(text):
    pos_word = kkma.pos(text)
    
    #stopwords=['하다', '은', '는', '이', '인', '로', '의', '등', '또', '보다']

    word_list = []
    for word, tag in pos_word:
        if tag in ['NNG', 'NR', 'SW']:
            #if word not in stopwords:
            word_list.append(word)
    return word_list
```

```python
result = extract_pos(text)
result
```
