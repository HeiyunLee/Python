```python
rating_list = []
review_list = []
date_list = []

while True:
    # while은 반복횟수를 알 수 없을 때 사용한다.
    try:
        soup = bs(driver.page_source, 'lxml')
        time.sleep(1)

        rating = soup.find_all('i', {'data-hook':'review-star-rating'})
        review = soup.find_all('span', {'data-hook':'review-body'})
        date = soup.find_all('span', {'data-hook':'review-date'})

        for j in range(len(review)):
            rating_list.append(rating[j].text)
            review_list.append(review[j].text)
            date_list.append(date[j].text)
            print(rating[j].text, review[j].text, date[j].text)

        time.sleep(1)

        next_page = driver.find_element_by_css_selector('#cm_cr-pagination_bar > ul > li.a-last > a')
        next_page.click()

    except:
        break
```
