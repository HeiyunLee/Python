```python
import seaborn as sns     # 시각화하기 좋은 도구 seaborn
import matplotlib.pyplot as plt
```

```python
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")       # 글씨를 미세하게 선명하게 해준다.
#sns.set(font="Malgun Gothic")
#retina를 설정 -> 폰트 설정 지우기
```

```python
# 연도별 추계인구 수 변화추이
plt.figure(figsize=(15,4))
plt.xticks(rotation=90)     # 연도가 겹쳐보이기 때문에 로테이션으로 각도 조절해 연도 보이기
sns.pointplot(data=df, x="연도 (년)", y="추계인구(명)")
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=90)    
sns.lineplot(data=df, x="연도 (년)", y="추계인구(명)")
# lineplot 으로 그래프를 그리면 연도를 구간으로 묶어서 그래프를 보여준다.
```

```python
df
```

```python
# 연도별 출생자 수 변화추이
plt.figure(figsize=(15,5))
plt.xticks(rotation=90)     
sns.pointplot(data=df, x="연도 (년)", y="출생자수(명)")
```

```python
# 그래프 왼쪽 상단의 1e6 = 10^6(10의 6제곱)
(10**6) * 0.26  # -> 2021년도의 출생자수가 약 20만 명대 임을 계산할 수 있다.
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df, x="연도 (년)", y="출생자수(명)")
```

그래프를 보면 출생자 수가 점점 감소하는 것을 알 수 있다.

```python
# 연도별 사망자수 변화 추이
plt.figure(figsize=(15,5))
plt.xticks(rotation=60)
sns.pointplot(data=df, x="연도 (년)", y="사망자수(명)")
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df, x="연도 (년)", y="사망자수(명)")
```

- 전쟁시기에 사망자 수가 급증한 것을 확인할 수 있다.
- 시간이 지날수록 사망자 수가 증가하는데 총인구 수가 반영된 결과로 보인다.

```python
# seaborn을 사용해 그래프 그리기
plt.figure(figsize=(20,8))
plt.xticks(rotation=60)
sns.pointplot(data=df, x="연도 (년)", y="출생자수(명)")
sns.pointplot(data=df, x="연도 (년)", y="사망자수(명)", color="orange")
```

```python
df.columns    # 컬럼만 뺴서 보기
```

```python
# pandas로 그래프 그리기  -> plot()
# 원하는 컬럼만 가져오기 -> df[]  대괄호 사용
# 여러개의 컬럼을 사용할 떄는 list형태로 지정해줘야 함. -> df[[]]  -> 대괄호 두 개 사용..?
df_pop = df[['연도 (년)', '출생자수(명)', '사망자수(명)']]   # -> 연도를 x값으로 사용하기 위해 연도를 인덱스로 설정해 준다.
df_pop = df_pop.set_index("연도 (년)")    # 바인딩 해주기
df_pop.head()
```

연도가 인덱스로 바뀐 것을 확일할 수 있다.

```python
df_pop.plot(figsize=(15,5))
```

2019년에 출생자수와 사망자수가 같아지고 그 이후 사망자수가 출생자수보다 많아진다.

```python
# 최근 50년의 데이터만 보기 -> 슬라이싱
df_pop[-50:]
```
