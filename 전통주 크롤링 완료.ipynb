```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as req
import time
from tqdm import tqdm_notebook
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://thesool.com/front/find/M000000082/list.do")
```

```python
# 리스트 만들기
a_list = []
b_list = []
c_list = []
d_list = []
e_list = []
f_list = []
g_list = []
h_list = []
n_list = []
j_list = []
k_list = []
l_list = []
m_list = []
img_url = []
tag_list = []

while True:
    try:
        for i in range(1,72):
            # 1페이지 부터 71페이지까지 클릭
            page = driver.find_element_by_css_selector("a[onclick^='fn_egov_link_page({}); return false;']".format(i))
            page.click()
            time.sleep(1)
            
            for i in range(10):
                # 품목 이미지 클릭
                img = driver.find_elements_by_css_selector('div > span > img')
                img[i].click()
                time.sleep(1)
                
                # 품명
                a = driver.find_elements_by_class_name('subject')
                for i in a :
                    a_list.append(i.text)
                
                # 종류
                b = driver.find_elements_by_css_selector('dd.detail > div.info > ul > li:nth-child(1) > span')
                for i in b :
                    b_list.append(i.text)
                
                # 원재료
                c = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(2) > span')
                for i in c :
                    c_list.append(i.text)

                # 알콜도수
                d = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(3) > span')
                for i in d :
                    d_list.append(i.text)

                # 용량
                e = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(4) > span')
                for i in e :
                    e_list.append(i.text)
                
                # 수상내역 
                f = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(5) > span')
                for i in f :
                    f_list.append(i.text)
                
                # 기타
                g = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.detail > div.info > ul > li:nth-child(6) > span')
                for i in g :
                    g_list.append(i.text)
                
                # 제품소개 
                h = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.intro > div.text')
                for i in h :
                    h_list.append(i.text)
                
                # 어울리는 음식
                n = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.food > div.text')
                for i in n :
                    n_list.append(i.text)
                
                # 양조장명
                j = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(1) > span')
                for i in j :
                    j_list.append(i.text)
                
                # 주소
                k = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(2) > span')
                for i in k :
                    k_list.append(i.text)
                
                # 홈페이지 
                l = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(3) > span > a')
                for i in l :
                    l_list.append(i.text)
                
                # 문의
                m = driver.find_elements_by_css_selector('#content > div > div > div.product-view > dl > dd.place > div.text > ul > li:nth-child(4)')
                for i in m :
                    m_list.append(i.text)
                
                # 이미지 src                    
                images = driver.find_elements_by_css_selector("#content > div > div > div.product-view > dl > dd.detail > div.thumb > div > span > img")
                for i in images :
                    url = i.get_attribute('src')
                    img_url.append(url)
                
                # 태그                    
                tag = driver.find_elements_by_class_name('tag')
                for i in tag :
                    tag_list.append(i.text)
                
                # 뒤로가기 (목록으로) 클릭
                button = driver.find_element_by_class_name('btn-primary')
                button.click()
                time.sleep(1)
        
    except:
        break



# csv 파일로 만들기
df.to_csv("data.csv", mode='w')
```
