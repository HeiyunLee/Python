### 초음파 광물 예측하기: k겹 교차 검증

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
```

```python
# 깃허브에 준비된 데이터 가져오기
!git clone https://github.com/taehojo/data.git
```

```python
# 광물 데이터 불러오기
df = pd.read_csv('./data/sonar3.csv', header=None)
```

```python
# X: 음파 관련 속성, y: 광물 속성
X = df.iloc[:, 0:60]
y = df.iloc[:,60]
```

```python
# 몇 겹으로 나눌지 k값 설정
k = 5
```

```python
# kfold 함수 호출, 분할하기 전에 샘플이 치우치지 않도록 섞어 준다.(-> shuffle)
kfold = KFold(n_splits=k, shuffle=True)
```

```python
# 정확도가 채워질 빈 리스트 생성
acc_score = []
```

```python
def model_fn():
    model = Sequential()   # 딥러닝 모델의 구조 시작
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    return model
```

```python
# k겹 교차 검증을 이용해 k번의 학습 실행
# for 문에 의해 k번 반복
# split() 함구에 의해 k개의 학습셋, 테스트셋 분리

for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:] 
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model = model_fn()
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=False)
    
    accuracy = model.evaluate(X_test, y_test)[1]   # 정확도 구하기
    acc_score.append(accuracy)                     # 정확도 리스트에 저장
    
    
# k번 실시된 정확도의 평균 구하기
avg_acc_score = sum(acc_score)/k

# 결과 출력
print('장확도: ', acc_score)
print('정확도 평균: ', avg_acc_score)
```
